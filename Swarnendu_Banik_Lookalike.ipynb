```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
```

```python
customers_df = pd.read_csv('Customers.csv')
transactions_df = pd.read_csv('Transactions.csv')
products_df = pd.read_csv('Products.csv')
```

```python
transactions_with_products = transactions_df.merge(products_df, on='ProductID')
merged_data = transactions_with_products.merge(customers_df, on='CustomerID')

print(merged_data.columns)
```

```python
transactions_with_products = transactions_df.merge(
    products_df.rename(columns={'Price': 'ProductPrice'}), on='ProductID'
)

merged_data = transactions_with_products.merge(customers_df, on='CustomerID')
```

```python
customer_features = merged_data.groupby('CustomerID').agg(
    total_spending=('TotalValue', 'sum'),
    num_transactions=('TransactionID', 'count'),
    avg_price_per_transaction=('ProductPrice', 'mean'),
    most_common_category=('Category', lambda x: x.mode()[0])
).reset_index()
```

```python
encoded_customers = customers_df[['CustomerID', 'Region']].merge(customer_features, on='CustomerID')
encoded_customers = pd.get_dummies(encoded_customers, columns=['Region', 'most_common_category'], drop_first=True)
```

```python
scaler = StandardScaler()
features = scaler.fit_transform(encoded_customers.drop(columns=['CustomerID']))
```

```python
similarity_matrix = cosine_similarity(features)
```

```python
customer_ids = encoded_customers['CustomerID'].tolist()
lookalike_map = {}
```

```python
for i, customer_id in enumerate(customer_ids[:20]):  # First 20 customers
    similar_indices = similarity_matrix[i].argsort()[::-1][1:4]  # Top 3 most similar (exclude self)
    lookalikes = [(customer_ids[j], float(similarity_matrix[i][j])) for j in similar_indices]
    lookalike_map[customer_id] = lookalikes
```

```python
lookalike_df = pd.DataFrame({
    'CustomerID': lookalike_map.keys(),
    'Lookalikes': [str(lookalikes) for lookalikes in lookalike_map.values()]
})
lookalike_df.to_csv('Swarnendu_Banik_Lookalike.csv', index=False)
```
