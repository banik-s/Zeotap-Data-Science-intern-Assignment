```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
customers_df = pd.read_csv('Customers.csv')
transactions_df = pd.read_csv('Transactions.csv')
```

```python
transactions_with_customers = transactions_df.merge(customers_df, on='CustomerID')

transactions_with_customers.sample(10)
```

```python
customer_transactions = transactions_with_customers.groupby('CustomerID').agg(
    total_spending=('TotalValue', 'sum'),
    num_transactions=('TransactionID', 'count'),
    avg_spending=('TotalValue', 'mean')
).reset_index()
```

```python
customer_transactions.head()
```

```python
merged_data = customers_df.merge(customer_transactions, on='CustomerID', how='left')
```

```python
merged_data.head()
```

```python
merged_data.fillna({'total_spending': 0, 'num_transactions': 0, 'avg_spending': 0}, inplace=True)
```

```python
merged_data.sample(10)
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_region = encoder.fit_transform(merged_data[['Region']])
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(merged_data[['total_spending', 'num_transactions', 'avg_spending']])
```

### Combine all features for clustering

```python
clustering_features = pd.DataFrame(scaled_features, columns=['total_spending', 'num_transactions', 'avg_spending'])
clustering_features = pd.concat([clustering_features, pd.DataFrame(encoded_region)], axis=1)
```

```python
db_indices = []
cluster_range = range(2, 11)
```

```python
clustering_features.columns = clustering_features.columns.astype(str)
```

```python
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(clustering_features)
    db_index = davies_bouldin_score(clustering_features, clusters)
    db_indices.append((k, db_index))
```

```python
optimal_k = min(db_indices, key=lambda x: x[1])
```

```python
final_kmeans = KMeans(n_clusters=optimal_k[0], random_state=42, n_init=10)    # Using K-Means clustering
merged_data['Cluster'] = final_kmeans.fit_predict(clustering_features)
```

```python
plt.figure(figsize=(8, 5))
plt.plot([k[0] for k in db_indices], [k[1] for k in db_indices], marker='o')
plt.title('DB Index for Different Cluster Numbers')
plt.xlabel('Number of Clusters')
plt.ylabel('DB Index')
plt.grid()
plt.show()
```

```python
pca = PCA(n_components=2)
pca_features = pca.fit_transform(clustering_features)
plt.figure(figsize=(10, 6))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=merged_data['Cluster'], cmap='viridis', s=50)
plt.title('Customer Clusters Visualized with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
merged_data.to_csv('Clustered_Customers.csv', index=False)
```
